```python
data.dropna()  # Drop rows with missing values
data['director'] = data['director'].fillna("No Data Available")
data['cast'] = data['cast'].fillna("No Data Available")
data['country'] = data['country'].fillna("No Data Available")
data['date_added']=data['date_added'].fillna(data['date_added'].mode()[0])          
data['rating']=data['rating'].fillna(data['rating'].mode()[0])      
data['duration']=data['duration'].fillna('No Data Available')  
data
```

Irregular Data

Detect outliers in your dataset and handle them as needed. Use code comments to make notes about your thought process.

```python
df=pd.read_csv('netflix_titles.csv')

features = df[['release_year']]

model = IsolationForest(contamination=0.05, random_state=42)

outliers = model.fit_predict(features)

df['outlier'] = outliers

plt.scatter(df['release_year'], np.zeros_like(df['release_year']), c=df['outlier'], cmap='viridis')
plt.xlabel('Release Year')
plt.title('Outlier Detection in Dataset')
plt.show()

outlier_df = df[df['outlier'] == -1]
print("Outliers:")
print(outlier_df)
```
